```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By

```

```python
r=webdriver.Chrome()
url="https://www.imdb.com/chart/top/"
r.get(url)
```

```python
name=r.find_elements("xpath","//div[@class='ipc-title ipc-title--base ipc-title--title ipc-title-link-no-icon ipc-title--on-textPrimary sc-b189961a-9 bnSrml cli-title']/a/h3")
year=r.find_elements("xpath","//div[@class='sc-b189961a-0 iqHBGn cli-children']/div[@class='sc-b189961a-7 btCcOY cli-title-metadata']/span[1]")
rating=r.find_elements("xpath","//span[@class='ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating']/span[@class='ipc-rating-star--rating']")
view=r.find_elements("xpath","//span[@class='ipc-rating-star ipc-rating-star--base ipc-rating-star--imdb ratingGroup--imdb-rating']/span[@class='ipc-rating-star--voteCount']")
poster_link=r.find_elements("xpath","//div[@class='ipc-poster ipc-poster--base ipc-poster--dynamic-width ipc-sub-grid-item ipc-sub-grid-item--span-2']/div/img")
```

```python
movie_details=[]
for i in range(len(name)):
    rank=i+1
    Name=name[i].text
    Year=year[i].text
    # NEW_YEAR=Year.split(" ")[0]
    Rating=rating[i].text
    View=view[i].text
    Poster=poster_link[i].get_attribute("src")

    movie_details.append({
        "Rank":rank,
        "Movie_name":Name,
        "Release_Year":Year,
        "Rating_movie":Rating,
        "View_count":View,
        "Poster_link":Poster
    })
df=pd.DataFrame(movie_details)  
print(df.to_string())
df.to_csv("movies_details.csv",index=False)



```
